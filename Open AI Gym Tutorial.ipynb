```python
'''
    A simple implementation of the cartpole problem, using Q-learning. We will cover this in detail later in the 
    text. This can be treated as an example of how
    environments work
'''
import gym
env = gym.make('CartPole-v0')
env.reset()
for _ in range(1000):
    env.render()
    env.step(env.action_space.sample())
env.close()
```

```python
'''
    Another environment
'''
env = gym.make('MountainCar-v0')
env.reset()
for _ in range(1000):
    env.render()
    env.step(env.action_space.sample())
env.close()
```

```python
'''
    If we ever want to do better than take random actions at each step, it’d probably be good to actually know what 
    our actions are doing to the environment. We can use the step function for this. Step function will give us observations,
    reward, done and info. 
    
    https://gym.openai.com/docs/
    
'''
env = gym.make('CartPole-v0')
for i_episode in range(20):
    observation = env.reset()
    for t in range (100):
        env.render()
        #print(observation)
        action = env.action_space.sample()
        '''
            Observation seems to be list of 4 nos for each time step. Looks like it could be the position velocity etc
            Done is simply a bolean that tells us of the episode is over
            Reward is always 1.0 at every time step
            Info is an empty dictionary
        '''
        observation, reward, done, info = env.step(action)
        print(observation)
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
    
```

```python
'''
    Spaces
    
    In the examples above, we’ve been sampling random actions from the environment’s action space. But what actually 
    are those actions? Every environment comes with an action_space and an observation_space. These attributes are of 
    type Space, and they describe the format of valid actions and observations:
'''

import gym
env = gym.make('CartPole-v0')
print(env.action_space)
print(env.observation_space)
```

The Discrete space allows a fixed range of non-negative numbers, so in this case valid actions are either 0 or 1. The Box space represents an n-dimensional box, so valid observations will be an array of 4 numbers.

```python
'''
    We can also check the limits of the boxes. This does not work for discrete
'''
print(env.observation_space.high)
print(env.observation_space.low)
```
